```python
# 4.1 Perso mais pas fini
import numpy as np

def sigmoid(a):
    return 1/(1+np.e**(-a))

def relu(a):
    return max(0, a)

def forward_propagation(x, bias_hid, weight_hid, bias_out, weight_out):
    # hidden layer computation
    a = np.dot(weight_hid.T, x) + bias_hid
    # f(a) for hidden layer
    for i in range(len(a)):
        for j in range(len(a[i])):
            a[i][j] = relu(a[i][j])
    h = a
    # output layer computation
    a = np.dot(weight_out.T, h) + bias_out
    # f(a) for output layer
    return sigmoid(a[0][0])

# J function (cost) = the Cross-Entropy loss function bc no regu

# For example from slide (slide 13): check h values
# bias = np.array([[0],[-1]])
# weight = np.array([[1, 1],[1, 1]])

x1 = np.array([[0],[0]])
x2 = np.array([[0],[1]])
x3 = np.array([[1],[0]])
x4 = np.array([[1],[1]])
X = [x1, x2, x3, x4]
bias = np.array([[0.7],[-1.10]])
weight = np.array([[0.83, 1.28],[0.91, 1.11]])
bias_out = -0.25
weight_out = np.array([[0.7],[-1.96]])
learn_rate = 0.25  # le n prolongé (éta?)
regularization = 0 # le lambda
print("Question 1:")
y_pred = []
for i in range(len(X)):
    yi = forward_propagation(X[i], bias, weight, bias_out, weight_out)
    y_pred.append(yi)
    print("x"+str(i+1)+":", yi)
print()
    
print("Question 2:")
print("Missclassified: x1 to 1\nWell classified x2 & x3 to 1, x4 to 0\nSo 3 examples well classified")
print()

print("Question 3:")
y_true = [0, 1, 1, 0]
def CE_loss(y_pred, y_true):
    summ = 0
    for i in range(len(y_pred)):
        summ -= y_true[i]*np.log(y_pred[i]) + (1-y_true[i])*np.log(1-y_pred[i])
    return summ
print(CE_loss(y_pred, y_true))
print()

print("Question 4:")
def out_layer_gradient_J(y_pred, y_true):
    """y_pred & y_true are only for a sample here (only number)
    lambda of regularization is equal to 0 here.
    """
    return -(y_true - y_pred) / (y_pred - y_pred**2)
print(out_layer_gradient_J(y_pred[0], y_true[0]))
print()

print("Question 5:")
def gradient_weight_J(y_pred, y_true, X):
    # Slide 9 ou slide 24
    ret = []
    for j in range(len(X[0])):
        summ = 0
        for i in range(len(X)):
            summ -= (y_true[i] - y_pred[i])*X[i][j]
        ret.append(summ[0])
    return ret
print("grad_w_1", gradient_weight_J(y_pred, y_true, X)[0])
print("grad_w_2", gradient_weight_J(y_pred, y_true, X)[1])
print("grad_b", out_layer_gradient_J(y_pred[0], y_true[0]))
print()

print("Question 6")

```
